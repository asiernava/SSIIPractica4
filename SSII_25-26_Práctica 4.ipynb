### Clase nodo

```python
class Node:
    def __init__(self, parent=None, state=None, cost=None, action=None, depth=0):
        # Nodo del √°rbol de b√∫squeda
        self.parent = parent       # Nodo padre
        self.state = state         # Estado del nodo (aqu√≠ ser√° la informaci√≥n del servidor)
        self.cost = cost           # Coste acumulado
        self.action = action       # Acci√≥n / regla aplicada
        self.depth = depth         # Profundidad del nodo

    def __str__(self):
        return str(self.state)

    # Setters y getters
    def set_parent(self, parent):
        self.parent = parent

    def get_parent(self):
        return self.parent

    def set_state(self, state):
        self.state = state

    def get_state(self):
        return self.state

    def set_cost(self, cost):
        self.cost = cost

    def get_cost(self):
        return self.cost

    def set_action(self, action):
        self.action = action

    def get_action(self):
        return self.action

    def set_depth(self, depth):
        self.depth = depth

    def get_depth(self):
        return self.depth
```

```python
class EstadoServidorPosicion:
    def __init__(self, posicion=None, matriz=None, acciones_realizadas=None):
        # Posici√≥n del servidor en el centro de datos (fila, columna)
        self.posicion = posicion
        
        # Matriz del data center (solo lectura, no se modifica)
        self.matriz = matriz
        
        # Lista de acciones realizadas para llegar aqu√≠ (opcional)
        self.acciones_realizadas = acciones_realizadas if acciones_realizadas is not None else []

    # -------- GETTERS --------
    def get_posicion(self):
        return self.posicion

    def get_matriz(self):
        return self.matriz

    def get_acciones(self):
        return self.acciones_realizadas

    # -------- SETTERS --------
    def set_posicion(self, nueva_posicion):
        self.posicion = nueva_posicion

    def set_acciones(self, acciones):
        self.acciones_realizadas = acciones

    def add_accion(self, accion):
        self.acciones_realizadas.append(accion)

    # -------- STRING --------
    def __str__(self):
        fila, col = self.posicion
        return f"Posici√≥n actual: ({fila}, {col}) - Servidor: {self.matriz[fila][col]}"
```

### Movimientos dentro de la matriz

```python
def mover_arriba(estado):
    fila, col = estado.get_posicion()
    matriz = estado.get_matriz()

    nueva_fila = fila - 1
    nueva_col = col

    # 1. Comprobar l√≠mites
    if nueva_fila < 0:
        return None

    # 2. Comprobar si hay obst√°culo
    if matriz[nueva_fila][nueva_col] == 'X':
        return None

    # 3. Devolver nuevo estado
    nuevo_estado = EstadoServidorPosicion(
        posicion=(nueva_fila, nueva_col),
        matriz=matriz
    )
    return nuevo_estado


def mover_abajo(estado):
    fila, col = estado.get_posicion()
    matriz = estado.get_matriz()

    nueva_fila = fila + 1
    nueva_col = col

    if nueva_fila >= len(matriz):
        return None

    if matriz[nueva_fila][nueva_col] == 'X':
        return None

    return EstadoServidorPosicion(
        posicion=(nueva_fila, nueva_col),
        matriz=matriz
    )


def mover_izquierda(estado):
    fila, col = estado.get_posicion()
    matriz = estado.get_matriz()

    nueva_fila = fila
    nueva_col = col - 1

    if nueva_col < 0:
        return None

    if matriz[nueva_fila][nueva_col] == 'X':
        return None

    return EstadoServidorPosicion(
        posicion=(nueva_fila, nueva_col),
        matriz=matriz
    )


def mover_derecha(estado):
    fila, col = estado.get_posicion()
    matriz = estado.get_matriz()

    nueva_fila = fila
    nueva_col = col + 1

    if nueva_col >= len(matriz[0]):
        return None

    if matriz[nueva_fila][nueva_col] == 'X':
        return None

    return EstadoServidorPosicion(
        posicion=(nueva_fila, nueva_col),
        matriz=matriz
    )
```

## 1. BFS

BFS es un algoritmo de b√∫squeda que recorre un grafo o √°rbol por niveles.
Primero explora todos los nodos m√°s cercanos al inicio antes de pasar a los siguientes.
Usa una cola para ir guardando los nodos que debe visitar y garantiza encontrar el camino m√°s corto si todos los pasos valen lo mismo.

## 1.1 BFS TREESEARCH

(Explicacion tree)

### 1.1.1 C√ìDIGO

```python
def treesearch_bfs(matriz, start, goal_number):  
    frontier = []  
    found = False  
    path = []  
    iteration_limit = 1000  
    iterations = 0  

    # Nodo inicial
    node1 = Node()
    node1.set_parent(None)
    node1.set_state(start)
    node1.set_action(get_neighbors(matriz, start, goal_number))
    node1.set_cost(0)
    node1.set_depth(0)
    frontier.append(node1)

    print("\n--- INICIO treesearch_bfs ---")

    while not found and frontier and iterations < iteration_limit:
        iterations += 1
        node_parent = frontier.pop(0)  
        current_state = node_parent.get_state()

        # Mostrar nodo expandido
        print(f"\nExpandiendo: {current_state}")
        if node_parent.get_parent():
            print(f"Parent: {node_parent.get_parent().get_state()}")
        print(f"Cost: {node_parent.get_cost()}")
        print(f"Action (Neighbors list): {node_parent.get_action()}")
        print(f"Depth: {node_parent.get_depth()}")
        print("*******************************")

        # Revisamos si el valor en la matriz coincide con la meta
        valor_actual = matriz[current_state[0]][current_state[1]][current_state[2]]
        if str(valor_actual) == str(goal_number):
            found = True
            node1 = node_parent
            break

        # Expandir y a√±adir hijos a la frontera
        next_nodes = expand(frontier, node_parent, matriz, goal_number)
        frontier.extend(next_nodes)

    if found:
        # Reconstruir camino desde el objetivo hasta el inicio
        total_cost = node1.get_cost()
        temp_node = node1
        while temp_node:
            path.append(temp_node.get_state())
            temp_node = temp_node.get_parent()
        path.reverse()

        print("\n¬°OBJETIVO ENCONTRADO!")
        print("El camino ha sido " + str(path))
        print(f"Total cost: {total_cost}")
        return node1
    else:
        print("solution not found")
        return None
```

GET_NEIGHBORS --> Devuelve las posiciones a las que se puede mover desde un nodo.
Es decir, mira la matriz y te dice qu√© casillas adyacentes son v√°lidas (no son ‚ÄòX‚Äô, est√°n dentro de los l√≠mites, etc.).

```python
def get_neighbors(matriz, position, goal_number):
    nivel, row, col = position
    rows = len(matriz[nivel])
    cols = len(matriz[nivel][0])
    neighbors = []

    goal_number = int(goal_number)
    target_level = 1 if goal_number > 99 else 0

    # -------------------------------------------------------------------
    # 1. SI ESTAMOS EN EL NIVEL EQUIVOCADO ‚Üí NOS MOVEREMOS NORMALMENTE
    #    HASTA LLEGAR A 'UP' o 'DOWN'
    # -------------------------------------------------------------------
    if nivel != target_level:

        # Movimientos normales buscando el portal (UP/DOWN)
        directions = [(0,1),(1,0),(0,-1),(-1,0)]
        for dr, dc in directions:
            new_row = row + dr
            new_col = col + dc
            if 0 <= new_row < rows and 0 <= new_col < cols:
                if matriz[nivel][new_row][new_col] != 'X':
                    neighbors.append((nivel, new_row, new_col))

        # Si estamos SOBRE el portal, a√±adir salto de nivel
        current_value = matriz[nivel][row][col]

        if current_value == 'UP' and nivel + 1 < len(matriz):
            neighbors.append((nivel + 1, row, col))

        if current_value == 'DOWN' and nivel - 1 >= 0:
            neighbors.append((nivel - 1, row, col))

        return neighbors

    # -------------------------------------------------------------------
    # 2. SI ESTAMOS YA EN EL NIVEL CORRECTO ‚Üí MOVER NORMALMENTE
    # -------------------------------------------------------------------
    directions = [(0,1),(1,0),(0,-1),(-1,0)]
    for dr, dc in directions:
        new_row = row + dr
        new_col = col + dc
        if 0 <= new_row < rows and 0 <= new_col < cols:
            if matriz[nivel][new_row][new_col] != 'X':
                neighbors.append((nivel, new_row, new_col))

    return neighbors
```

EXPAND --> Crea los nodos hijos usando los vecinos que devuelve get_neighbors.
Cada vecino se transforma en un nuevo Node con su padre, coste, profundidad, etc.

```python
def expand(frontier, node_parent, matriz, goal_number):
    children = []
    for neighbor in get_neighbors(matriz, node_parent.get_state(), goal_number):
        child = Node()
        child.set_state(neighbor)
        child.set_parent(node_parent)
        child.set_action(get_neighbors(matriz, neighbor, goal_number))
        child.set_cost(node_parent.get_cost() + 1)
        child.set_depth(node_parent.get_depth() + 1)
        children.append(child)
    return children
```

```python
def graphsearch_bfs(matriz, start, goal_number):  
    goal_number = str(goal_number)  # convertir goal a string
    frontier = []  
    found = False  
    path = []  
    iteration_limit = 1000  
    iterations = 0  

    explored = set()

    node1 = Node()
    node1.set_parent(None)
    node1.set_state(start)
    node1.set_action(get_neighbors(matriz, start, goal_number))
    node1.set_cost(0)
    node1.set_depth(0)
    frontier.append(node1)
    explored.add(start)

    print("\n--- INICIO graphsearch_bfs ---")

    while not found and frontier and iterations < iteration_limit:
        iterations += 1
        node_parent = frontier.pop(0)  
        current_state = node_parent.get_state()

        print(f"\nExpandiendo: {current_state}")
        if node_parent.get_parent():
            print(f"Parent: {node_parent.get_parent().get_state()}")
        print(f"Cost: {node_parent.get_cost()}")
        print(f"Action (Neighbors list): {node_parent.get_action()}")
        print(f"Depth: {node_parent.get_depth()}")
        print("*******************************")

        valor_actual = matriz[current_state[0]][current_state[1]][current_state[2]]
        if str(valor_actual) == goal_number:
            found = True
            node1 = node_parent
            break

        next_nodes = expand(frontier, node_parent, matriz, goal_number)

        for child in next_nodes:
            child_state = child.get_state()
            if child_state not in explored:
                explored.add(child_state)
                frontier.append(child)

        print(f"Frontier ({len(frontier)} nodos): {[n.get_state() for n in frontier]}")

    if found:
        total_cost = node1.get_cost()
        temp_node = node1
        while temp_node:
            path.append(temp_node.get_state())
            temp_node = temp_node.get_parent()
        path.reverse()

        print("\n¬°OBJETIVO ENCONTRADO!")
        print("El camino ha sido " + str(path))
        print(f"Total cost: {total_cost}")
        return node1
    else:
        print("solution not found")
        return None
```

GET_NEIGHBORS --> Devuelve las posiciones a las que se puede mover desde un nodo.
Es decir, mira la matriz y te dice qu√© casillas adyacentes son v√°lidas (no son ‚ÄòX‚Äô, est√°n dentro de los l√≠mites, etc.).

```python
def get_neighbors(matriz, position, goal_number):
    nivel, row, col = position
    rows = len(matriz[nivel])
    cols = len(matriz[nivel][0])
    neighbors = []

    goal_number = int(goal_number)
    target_level = 1 if goal_number > 99 else 0

    # -------------------------------------------------------------------
    # 1. SI ESTAMOS EN EL NIVEL EQUIVOCADO ‚Üí NOS MOVEREMOS NORMALMENTE
    #    HASTA LLEGAR A 'UP' o 'DOWN'
    # -------------------------------------------------------------------
    if nivel != target_level:

        # Movimientos normales buscando el portal (UP/DOWN)
        directions = [(0,1),(1,0),(0,-1),(-1,0)]
        for dr, dc in directions:
            new_row = row + dr
            new_col = col + dc
            if 0 <= new_row < rows and 0 <= new_col < cols:
                if matriz[nivel][new_row][new_col] != 'X':
                    neighbors.append((nivel, new_row, new_col))

        # Si estamos SOBRE el portal, a√±adir salto de nivel
        current_value = matriz[nivel][row][col]

        if current_value == 'UP' and nivel + 1 < len(matriz):
            neighbors.append((nivel + 1, row, col))

        if current_value == 'DOWN' and nivel - 1 >= 0:
            neighbors.append((nivel - 1, row, col))

        return neighbors

    # -------------------------------------------------------------------
    # 2. SI ESTAMOS YA EN EL NIVEL CORRECTO ‚Üí MOVER NORMALMENTE
    # -------------------------------------------------------------------
    directions = [(0,1),(1,0),(0,-1),(-1,0)]
    for dr, dc in directions:
        new_row = row + dr
        new_col = col + dc
        if 0 <= new_row < rows and 0 <= new_col < cols:
            if matriz[nivel][new_row][new_col] != 'X':
                neighbors.append((nivel, new_row, new_col))

    return neighbors
```

EXPAND --> Crea los nodos hijos usando los vecinos que devuelve get_neighbors.
Cada vecino se transforma en un nuevo Node con su padre, coste, profundidad, etc.

```python
def expand(frontier, node_parent, matriz, goal_number):
    children = []
    for neighbor in get_neighbors(matriz, node_parent.get_state(), goal_number):
        child = Node()
        child.set_state(neighbor)
        child.set_parent(node_parent)
        child.set_action(get_neighbors(matriz, neighbor, goal_number))
        child.set_cost(node_parent.get_cost() + 1)
        child.set_depth(node_parent.get_depth() + 1)
        children.append(child)
    return children
```

## DFS GRAPH

DFS es un algoritmo de b√∫squeda que recorre un grafo o √°rbol profundizando al m√°ximo antes de retroceder.
Primero baja por un camino hasta que no puede seguir (si no puede mas), y luego vuelve atr√°s para explorar otras opciones.
Usa una pila para recordar por d√≥nde continuar y no garantiza encontrar el camino m√°s corto, pero es √∫til para explorar estructuras profundas.

```python
def treesearch_dfs(matriz, start, goal): # Renombro 'graph' a 'matriz' para consistencia

  frontier = []  
  found = False  
  current_depth = 0  
  path = []  
  iteration_limit = 200 
  iterations = 0  
  
  # *** ADICI√ìN CLAVE PARA GRAPH SEARCH ***
  explored = set() 
  
  node_goal = None # Variable para almacenar el nodo objetivo

  node1 = Node()
  node1.set_parent(None)
  node1.set_state(start)
  node1.set_action(get_neighbors(matriz, start))
  node1.set_cost(0)
  node1.set_depth(current_depth)
  frontier.append(node1)
  
  # A√±adir el estado inicial a explorados
  explored.add(start)

  if node1.get_state() == goal:
    found = True
    return node1
  else:
    found = False
    
  print("\n--- INICIO treesearch_dfs (Graph Search) ---")


  while not found and frontier and iterations < iteration_limit:
    iterations += 1
    node_parent = frontier.pop() # LIFO (Profundidad)
    current_state = node_parent.get_state()

    # Requisito 10: Mostrar el nodo expandido
    print(f"\nExpandiendo: {current_state}")
    if node_parent.get_parent():
      print(f"Parent: {node_parent.get_parent().get_state()}")
    print(f"Cost: {node_parent.get_cost()}")
    print(f"Action (Neighbors list): {node_parent.get_action()}")
    print(f"Depth: {node_parent.get_depth()}")
    print("*******************************")


    if current_state == goal:
      found = True
      node_goal = node_parent # Guardamos el nodo meta
      break
    else:
      # Llama a expand
      next_Nodes = expand(frontier, node_parent, matriz)
      
      # Filtramos los sucesores para implementar Graph Search
      newly_added_nodes = []
      for next_node in next_Nodes:
          state_to_check = next_node.get_state()
          
          # Solo a√±adimos a la frontera si el estado es NUEVO
          if state_to_check not in explored:
              if state_to_check == goal:
                  found = True
                  node_goal = next_node
                  break # Salir del bucle interno si encontramos el objetivo
              
              explored.add(state_to_check)
              newly_added_nodes.append(next_node)
              
      # DFS a√±ade los nodos nuevos al final de la pila (LIFO)
      frontier.extend(newly_added_nodes)
      
      # Requisito 10: Mostrar la frontera actual
      if not found:
        print(f"Frontier ({len(frontier)} nodos): {[n.get_state() for n in frontier]}")
        

  # --- Resultado Final ---
  if found:
    total_cost = node_goal.get_cost() # Usar node_goal
    temp_node = node_goal
    while temp_node:
      path.append(temp_node.get_state())
      temp_node = temp_node.get_parent()
    path.reverse()
    
    print("\n¬°OBJETIVO ENCONTRADO!")
    print("El camino ha sido " + str(path))
    print(f"Total cost: {total_cost}")
    return node_goal
  else:
    print("solution not found")
    return None
```

```python
# Funci√≥n expandir
def expand(frontier, node, matriz):

  Sucessors = []
  acciones_disponibles = node.get_action()

  for element in acciones_disponibles:
      node_Nuevo = Node()
      node_Nuevo.set_parent(node)
      node_Nuevo.set_state(element)
      node_Nuevo.set_action(get_neighbors(matriz, element))
      node_Nuevo.set_cost(node.get_cost() + 1)
      node_Nuevo.set_depth(node.get_depth() + 1)
      Sucessors.append(node_Nuevo)
  return Sucessors
```

## GREEDY GRAPH

Definir una nueva clase nodo para greedy, que incorpore la heuristica.

```python
class Node:
    def __init__(self, parent=None, state=None, cost=None, action = None, depth=0, heuristic=None):
        self.parent = parent
        self.state = state
        self.cost = cost
        self.action = action
        self.depth = depth
        self.heuristic = heuristic

    def __str__(self):
        return self.state

    def set_parent(self, parent):
        self.parent = parent

    def get_parent(self):
        return self.parent

    def set_state(self, state):
        self.state = state

    def get_state(self):
        return self.state

    def set_cost(self, cost):
        self.cost = cost

    def get_cost(self):
        return self.cost

    def set_action(self, action):
        self.action = action

    def get_action(self):
        return self.action

    def set_depth(self, depth):
        self.depth = depth

    def get_depth(self):
        return self.depth

    def set_heuristic(self, heuristic):
        self.heuristic = heuristic

    def get_heuristic(self):
        return self.heuristic
```

Generar la heuristica.

```python
# Se genera en base a la distancia Manhattan --> https://interactivechaos.com/es/manual/tutorial-de-machine-learning/distancia-manhattan
def generar_matriz_heuristica(matriz, objetivo):
    rows, cols = len(matriz), len(matriz[0]) # Guardamos en dos variables el tama√±o de la matriz
    objetivo_x, objetivo_y = objetivo
    matriz_heuristica = []

    # Mediamte dos for (uno dentro del otro) creamos una nueva matriz con los nuevos valores
    for i in range(rows):
        fila_heuristica = []
        for j in range(cols):
            if matriz[i][j] == 'X': # Donde antes hab√≠a X, ahora tambi√©n
                fila_heuristica.append('X')
            else:
                distancia = abs(objetivo_x - i) + abs(objetivo_y - j) # M√©todos para que nos salgan los valores absolutos, si no falla
                fila_heuristica.append(distancia)
        matriz_heuristica.append(fila_heuristica)
    return matriz_heuristica
```

```python
# Heur√≠stica para la primera prueba (DESTINO_PRUEBA1 tiene el valor de 58)
matriz_heuristica = generar_matriz_heuristica(matriz, DESTINO_PRUEBA1)
for fila in matriz_heuristica:
    print(fila)
```

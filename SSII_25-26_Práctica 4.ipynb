```python
class Node:
    def __init__(self, parent=None, state=None, cost=None, action=None, depth=0):
        # Nodo del 谩rbol de b煤squeda
        self.parent = parent       # Nodo padre
        self.state = state         # Estado del nodo (aqu铆 ser谩 la informaci贸n del servidor)
        self.cost = cost           # Coste acumulado
        self.action = action       # Acci贸n / regla aplicada
        self.depth = depth         # Profundidad del nodo

    def __str__(self):
        return str(self.state)

    # Setters y getters
    def set_parent(self, parent):
        self.parent = parent

    def get_parent(self):
        return self.parent

    def set_state(self, state):
        self.state = state

    def get_state(self):
        return self.state

    def set_cost(self, cost):
        self.cost = cost

    def get_cost(self):
        return self.cost

    def set_action(self, action):
        self.action = action

    def get_action(self):
        return self.action

    def set_depth(self, depth):
        self.depth = depth

    def get_depth(self):
        return self.depth
```
